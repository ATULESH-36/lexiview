```python
import requests
import os
import sys
sys.path.append(os.path.abspath('..'))
from config.config import BASE_DIR
from src.utils import ensure_dir
```

```python
RAW_DATA_PATH=os.path.join(BASE_DIR,'data','raw')
os.makedirs(RAW_DATA_PATH,exist_ok=True)
```

```python
##fuunction to download data from url
def download_book(book_id,title):
    url=f"https://www.gutenberg.org/files/{book_id}/{book_id}-0.txt"
    response =requests.get(url)
    
    if response.status_code==200:
        file_path=os.path.join(RAW_DATA_PATH,f'{title}.txt')
        with open(file_path,'w',encoding='utf-8') as f:
            f.write(response.text)
        print(f"{title} downloaded successfully.")
    
    else:
        print(f"failed to download the book with id {book_id}.")        
        
    
```

```python
books_to_download={
    1342: "pride_and_prejudice",
    1400: "great_expectations",
    76: "huckleberry_finn",
    2701: "moby_dick",
    1524: "hamlet"
}
```

```python
for book_id,title in books_to_download.items():
    download_book(book_id,title)
```
