```python
import os
import sys
sys.path.append(os.path.abspath('..'))
from config.config import BASE_DIR
from src.utils import list_text_files
from src.utils import ensure_dir
```

```python
from src.preprocessing import preprocess_pipeline
```

```python
RAW_DATA_PATH=os.path.join(BASE_DIR,'data','raw')
CLEAN_DATA_PATH=os.path.join(BASE_DIR,'data','cleaned')
ensure_dir(CLEAN_DATA_PATH)
```

```python
for file in list_text_files(RAW_DATA_PATH):
        with open(os.path.join(RAW_DATA_PATH,file),'r',encoding='utf-8') as f:
             text=f.read()
        
        tokens=preprocess_pipeline(text)
        
        with open(os.path.join(CLEAN_DATA_PATH,file),'w',encoding='utf-8') as f:
            f.write(" ".join(tokens))  
        
        print(f"processed:{file}")    
```

In this phase I have removed the gutenberg header and footer,the illustration section,contents section ,all the digits and punctuation marks.All we are left with is only the contents of the chapters of a book.
