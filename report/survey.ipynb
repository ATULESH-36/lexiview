## 1.1 Introduction to Text Complexity

Text complexity analysis is an important area in Natural Language Processing (NLP) that focuses on evaluating how difficult a text is to read or understand. It plays a crucial role in applications such as readability assessment, authorship attribution, language modeling, and educational content design. One of the fundamental approaches to measuring text complexity is through lexical diversity, which reflects the richness and variety of vocabulary used in a text.

## 1.2 Lexical Diversity and TTR

Lexical diversity refers to the range of different words used in a text. A commonly used metric to quantify lexical diversity is the Type-Token Ratio(TTR),defined as the ratio of the unique words(type) to the total number of words (tokens). A higher TTR indicates a more diverse vocabulary, while a lower TTR suggests repetition and simpler language usage.

## 1.3 Limitations of TTR

Despite its simplicity, TTR suffers from a significant limitation:it is highly sensitive to text lenght. As the length  of a text increases, the likelihood of repeating words also increases which leands to a decrease in the TTR value. This makes it unreliable for comparing texts of different lengths, especially in large corpora such as literary works.

## 1.4 Improved Metrics

To address the limitations of the basic TTR, several improved lexical diversity metrics have been proposed. Root Type-Token Ratio (RTTR) normalizes TTR by dividing the number of types by the square root of the total tokens. Corrected Type-Token Ratio (CTTR) further refines this normalization by incorporating a scaling factor. Another advanced approach is the Moving Average Type-Token Ratio (MATTR), which computes TTR over a sliding window of fixed length across the text. MATTR reduces the bias introduced by text length and provides a more stable and reliable measure of lexical diversity.

## 1.5 Existing works

Lexical diversity measures have been widely used in various NLP applications. In authorship attribution, differences in vocabulary richness can help distinguish between writing styles of different authors. In genre classification, lexical diversity can reveal stylistic differences between narrative, scientific, and philosophical texts. Additionally, these metrics have been used to study language evolution by analyzing changes in vocabulary usage across time periods.

## 1.6 Gap Identified

Although multiple lexical diversity metrics exist, many studies rely heavily on the basic TTR without adequately addressing its limitations. There is a need for a comparative framework that evaluates multiple metrics, such as TTR, RTTR, CTTR, and MATTR, on real-world datasets. Furthermore, applying these metrics to large literary corpora can provide deeper insights into text complexity across authors, genres, and time periods.

## 1.8 Baseline Implementation of TTR

To understand the practical behavior of lexical diversity metrics, a basic implementation of the TTR is performed on a sample textual data. The implementation involved preprocessing the text by converting it to lowercase, removing punctuation, and tokenization it into individual words.The total numbers of tokens and unique word types were then computed to calculatethe TTR value.

This Preliminary experiment helped in observng the sensitivity of TTR to text length.It was noted that as the size of the text increased the TTR value tended to decrease due to prepition of words.This observation alogns with findings in existing literature and highlights the need for improved metrics such as MATTR. 

```python
import string

def preprocess_text(text):
    text=text.lower()
    print(text)
   
    text=text.translate(str.maketrans('','',string.punctuation)) ## this is used to remove the punctuation marks for a text.
    print(text)
    tokens=text.split()
    return tokens

def calculate_ttr(tokens):
    unique_tokens=len(set(tokens))
    print(f"unique tokens in the text  are:{unique_tokens}")
    total_tokens=len(tokens)
    print(f"total tokens in the text  are:{total_tokens}")
    if total_tokens==0:
        return 0
    ttr=unique_tokens/total_tokens
    return ttr
```

```python
text="This is just a simple implementation of what TTR is. This example is just for testing TTR."
tokens=preprocess_text(text)
ttr=calculate_ttr(tokens)

print(f"Type token ratio for the given text is :{ttr:.2f}")
```

The computed TTR value provides a basic estimate of vovabulary richness in the given text. However, this implementation also reveals that repeated words significantly reduce the TTR value,especially in longer texts. This reinforces the limitation of TTR and motivates the need for more robust metrics,which will be explored in the subsequent phases of this project.
